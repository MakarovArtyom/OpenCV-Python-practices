### Image Pyramids

At each level of **Image pyramid** we downsample our original image.<br>
In this notebook we are going to look throw the process of downsampling and construction of few layers of an image pyramid.

### Libraries import 

```python
import numpy as np
import cv2
import matplotlib.pyplot as plt 


%matplotlib inline 
```

Start with original image loading - we are going to display **rainbow image**.

```python
img = cv2.imread('images/rainbow_flag.jpg')
img_copy = np.copy(img)
img_copy = cv2.cvtColor(img_copy, cv2.COLOR_BGR2RGB)

f = plt.figure(figsize = (16,6))
plt.imshow(img_copy)
```

In the image pyramid the higher the layer the smaller the size of image. <br>
Let's use `cv2.pyrDown()`, that **inputs an original image (no need to convert to gray/binary) and returns a downsampled version of it**.<br>
Note, that function downlamples the image by **factor of 2**.

In this example, we are going to display **4 layers (original + 3 downsampled image) pyramid**.<br>
Note as well, that we are going to **specify x-axis and y-axis** in the way that:
 - images of all levels will lay on original-image axises. 
 - **x-axis**: from 0 to width of original image;
 - **y-axis**: from height of original image to 0.
 

```python
img_copy.shape
```

```python
level_1 = cv2.pyrDown(img_copy)
level_2 = cv2.pyrDown(level_1)
level_3 = cv2.pyrDown(level_2)
```

```python
f, (ax1, ax2, ax3, ax4) = plt.subplots(1,4, figsize = (20,7))

ax1.imshow(img_copy)
ax1.set_title('Original image')

ax2.imshow(level_1)
ax2.set_title('Level 1 image')
ax2.set_xlim(0, img_copy.shape[1])
ax2.set_ylim(img_copy.shape[0], 0)

ax3.imshow(level_2)
ax3.set_title('Level 2 image')
ax3.set_xlim(0, img_copy.shape[1])
ax3.set_ylim(img_copy.shape[0], 0)

ax4.imshow(level_3)
ax4.set_title('Level 3 image')
ax4.set_xlim(0, img_copy.shape[1])
ax4.set_ylim(img_copy.shape[0], 0)
```

As we can see from above result - our image was downsampled by factor of 2 on each level.<br>
Level 3 has a size of just **1/8 of original image**.
